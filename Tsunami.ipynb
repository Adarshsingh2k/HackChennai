```python
data=df.copy()
```

```python
data.isnull().sum()
```

```python
temp=data['LOCATION_NAME'].value_counts()
```

```python
data['LOCATION_FREQ']=data['LOCATION_NAME'].apply(lambda x: temp[x])
```

```python
for i in range(0, len(data)):
    if data['LOCATION_FREQ'][i] < 5:
        data['LOCATION_NAME'][i] = 'other'
```

```python
data1=data.drop(['LOCATION_NAME','REGION','COUNTRY','LOCATION_FREQ'],axis=1)
```

```python
data1['EVENT_VALIDITY'].replace({'Definite Tsunami':1,'Probable Tsunami':1,'Questionable Tsunami':1,'Very Doubtful Tsunami':0,
                                 'Event that only caused a seiche or disturbance in an inland river':0},inplace=True)
```

```python
data1['CAUSE'].value_counts()
```

```python
data1=data1[data1['TS_INTENSITY']>-1.5]
```

```python
data1=data1[data1['TS_INTENSITY']<7]
```

```python
data1['TS_INTENSITY'].value_counts()
```

```python
sns.boxplot(x=data1['TS_INTENSITY'])
```

```python
data1=pd.get_dummies(data1)
```

```python
data1
```

```python
new_data=data1.sample(frac=1).reset_index(drop=True)
```

```python
new_data
```

```python
x=new_data.drop('EVENT_VALIDITY',axis=1)
y=new_data['EVENT_VALIDITY']
```

```python
from sklearn.model_selection import train_test_split as tts
train_x,test_x,train_y,test_y=tts(x,y,stratify=y,random_state=10,test_size=0.25)
```

```python
columns=x.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(train_x)
train_scaled_x=scaler.transform(train_x)
test_scaled_x=scaler.transform(test_x)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,f1_score
```

```python
lr=LogisticRegression()
```

```python
lr.fit(train_scaled_x,train_y)
```

```python
k=lr.predict(test_scaled_x)
```

```python
b=accuracy_score(test_y,k)
b
```

```python
b=f1_score(test_y,k)
b
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
```

```python
from sklearn.model_selection import cross_val_score
```

```python
skfold = StratifiedKFold(n_splits=3,random_state=0)
model_c = RandomForestClassifier(criterion='entropy',n_estimators=10,random_state=0,n_jobs=-1)
results_skfold = cross_val_score(model_c, x, y.ravel(), cv=skfold)
print("Accuracy: %.2f%%" % (results_skfold.mean()*100.0))
```
