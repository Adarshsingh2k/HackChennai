```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw_data=pd.read_csv('eruptions.csv')
```

```python
data=raw_data.drop(['volcano_number', 'volcano_name', 'eruption_number','area_of_activity','evidence_method_dating'
                   ,'start_year','start_day','evidence_method_dating','end_year','end_day','end_month'],axis=1)
```

```python
data.isnull().sum()
```

```python
data['vei'].fillna(data['vei'].mode()[0],inplace=True)
```

```python
data['start_month'].fillna(data['start_month'].mode()[0],inplace=True)
```

```python
data
```

```python
data['eruption_category'].replace({'Confirmed Eruption':1,'Uncertain Eruption':0,'Discredited Eruption':0},inplace=True)
```

```python
new_data=data.sample(frac=1).reset_index(drop=True)
```

```python
new_data
```

```python
x=new_data.drop(['eruption_category'],axis=1)
y=new_data['eruption_category']
```

```python
from sklearn.model_selection import train_test_split as tts
train_x,test_x,train_y,test_y=tts(x,y,stratify=y,random_state=10,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(train_x)
```

```python
train_x=scaler.transform(train_x)
test_x=scaler.transform(test_x)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(train_x,train_y)
```

```python
lr.score(train_x,train_y)
```

```python
lr.score(test_x,test_y)
```

```python
from sklearn.ensemble import RandomForestClassifier
reg=RandomForestClassifier()
reg.fit(train_x,train_y)
```

```python
reg.score(train_x,train_y)
```

```python
reg.score(test_x,test_y)
```
