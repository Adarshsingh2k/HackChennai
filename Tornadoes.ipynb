```python
data= raw_data.drop(['om','date','time','yr','mo','st','dy','tz','stf','stn','fc','closs'],axis=1)
```

```python
data1=data.sample(frac=1).reset_index(drop=True)
```

```python
x=data1.drop(['mag'],axis=1)
y=data1['mag']
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
train_x,test_x,train_y,test_y=tts(x,y,random_state=10,stratify=y,test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(train_x)
scaler.transform(train_x)
scaler.transform(test_x)
```

```python
lr=LinearRegression()
lr.fit(train_x,train_y)
```

```python
lr.score(test_x,test_y)
```

```python
reg = RandomForestRegressor(random_state=56)
reg.fit(train_x, train_y)
k=reg.predict(test_x)
```

```python
r2_score(test_y,k)
```
