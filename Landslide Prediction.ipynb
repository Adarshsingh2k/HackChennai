```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
raw_data=pd.read_csv('catalog.csv')
```

```python
raw_data.describe(include='all')
```

```python
raw_data['trigger'].value_counts()
```

```python
raw_data['landslide_type'].value_counts()
```

```python
data=raw_data.drop(['geolocation','location_description','storm_name','source_name','source_link'
                    ,'hazard_type','id','date','time','continent_code','country_name','country_code',
                    'state/province','city/town'],axis=1)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data['distance'].fillna(data['distance'].mean(),inplace=True)
```

```python
data['latitude'].fillna(data['latitude'].mean(),inplace=True)
```

```python
data['longitude'].fillna(data['longitude'].mean(),inplace=True)
```

```python
data['landslide_type'].fillna(data['landslide_type'].mode()[0],inplace=True)
```

```python
data['landslide_size'].fillna(data['landslide_size'].mode()[0],inplace=True)
```

```python
data['trigger'].fillna(data['trigger'].mode()[0],inplace=True)
```

```python
data['injuries'].fillna(data['injuries'].mode()[0],inplace=True)
```

```python
data['fatalities'].fillna(data['fatalities'].mode()[0],inplace=True)
```

```python
data['trigger'].value_counts()
```

```python
data['trigger'].replace({'downpour':'Downpour','unknown':'Unknown','Continuous rain':'rain'},inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
  
# label_encoder object knows how to understand word labels. 
label_encoder =LabelEncoder() 
  
# Encode labels in column 'species'. 
data['trigger']= label_encoder.fit_transform(data['trigger']) 
```

```python
data['landslide_size'].replace({'medium':'Medium','small':'Small','large':'Large'},inplace=True)
```

```python
data['landslide_size'].value_counts()
```

```python
data['landslide_size'].replace({'Medium':1,'Small':0,'Large':2,'Very_large':3},inplace=True)
```

```python
data['landslide_size'].value_counts()
```

```python
data['landslide_type'].replace({'mudslide':'Mudslide','landslide':'Landslide','Rockslide':'Rockfall'},inplace=True)
```

```python
data['landslide_type']=label_encoder.fit_transform(data['landslide_type'])
```

```python
new_data=data.sample(frac=1).reset_index(drop=True)
```

```python
x=data.drop(['distance'],axis=1)
y=data['distance']
```

```python
new_data
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
train_x,test_x,train_y,test_y=tts(x,y,random_state=10,test_size=0.25)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
reg = RandomForestRegressor(random_state=56)
reg.fit(train_x, train_y)
k=reg.predict(test_x)
```

```python
reg.score(test_x,test_y)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(test_y,k)
```
